# Princípios (12)

1. Nossa maior prioridade é satisfazer o cliente, através da entrega adiantada e contínua de produto de valor.

2. Aceitar mudanças de requisitos, mesmo no fim do desenvolvimento.
Processos ágeis se adequam a mudanças, para que o cliente possa obter vantagens competitivas.

3. Entregar produtos funcionais com frequência, na escala de algumas semanas ou meses, de preferência com os períodos mais curtos.

4. Pessoas relacionadas a negócios e desenvolvedores devem trabalhar em conjunto e diariamente, durante todo o curso do projeto.

5. Construir projetos ao redor de indivíduos motivados.
Dar a eles ambiente e suporte necessários, e confiar que farão seu trabalho.

6. O método mais eficiente e eficaz de transmitir informações para, e por dentro de um time de desenvolvimento, é através de uma conversa cara a cara.

7. Produto funcional é a medida primária de progresso.

8. Processos ágeis promovem um ambiente sustentável.
Os patrocinadores, desenvolvedores e usuários devem ser capazes de manter indefinidamente passos constantes.

9. Contínua atenção à excelência técnica e bom design, aumenta a agilidade.

10. Simplicidade: a arte de maximizar a quantidade de trabalho que não precisou ser feito é essencial.

11. As melhores arquiteturas, requisitos e designs emergem de times auto-organizáveis.

12. Em intervalos regulares, o time reflete como ficar mais efetivo, então, se ajusta e otimiza seu comportamento com esse fim.

# Scrum

Fonte: https://www.youtube.com/watch?v=XfvQWnRgxG0

* framework de gerenciamento de projetos e desenvolvimento ágil
* não é processo detalhado e prescritivo
* base: desenvolvimento iterativo e incremental
* única equipe gerencia o desenvolvimento de um produto com entregas incrementais por ciclos iterativos (sprints)
* executado em timeboxes com duração constante (< 1 mês)


3 Pilares
1. Transparência
 * processos, requisitos de entrega e status. Ex: Kanban
2. Inspeção
 * constante de tudo que está sendo feito
3. Adaptação
 * de processo e produto às mudanças

Práticas Fundamentais

* Papéis
 1. Scrum Master (líder servidor, dono do processo)

 Ajudar a todos envolvidos (coach), facilitador

 * orientar como ou remover impedimentos e facilitar eventos
 * garantir que equipe compreenda, adote e siga regras do scrum


 2. Product Owner (PO: dono do produto)

 Ponto central, liderança, ordem, manter e comunicar

 * elabora visão do projeto
 * define e gerencia o backlog do produto
 * prioriza, maximiza e garante valor, entendimento e visibilidade do produto

 3. Dev Team

 * equipe auto-organizada e multifuncional
  * desenvolvedores
  * dba's
  * analistas
 * possuem o necessário para entregar incrementos de um produto funcional ao final de cada sprint sem dependência externa
 * gerenciam o backlog de atividades (sprint backlog) 
 * Construir projeto
 * 3-9 pessoas


* Eventos
1. Sprint Planning (ready)
  * reunião para criar backlog da sprint (o que fazer)

2. Daily Scrum
  * reunião 15 min diários do dev team (acompanhar atividades)
  1. o que fez ontem?
  2. o que vai fazer hoje?
  3. tem algum impedimento?

3. Sprint Execution
 * ciclo de desenvolvimento e uma iteração
 * com produto incremental e funcional ao término
4. Sprint Review (done)
 * reunião para demonstrar funcionalidades desenvolvidas na sprint
 * Surgem mudanças
 * Atualiza product backlog

5. Sprint Retrospective
 * Reunião Lições aprendidas

* Artefatos

   * Funcionalidades
   * lista ordenada e priorizada dos requisitos necessários
   * mantida, refinada (backlog refinement, grooming) pelo PO

3. Incremento/Entrega

# Lean Manufacturing – Sistema Toyota de Produção

fonte: https://annelisegripp.com.br/lean-o-que-essa-palavra-significa/


* Taiichi Ohno
* Japão pós 2a guerra
* Guerra Mundial
* Reduzir desperdício
* Mais qualidade nos produtos e serviços desenvolvidos
* Melhor atender o cliente
* Entrega contínua, enxuta, no menor tempo e com qualidade
1. Just-In-Time (JIT)
 * não produzir, transportar ou comprar antes de vender ou encomendar
2. Jidoka
 * automação com inteligência humana
 * fornece a máquinas e operadores
 * capacidade de detectar um problema que ocorreu e
 * interromper imediatamente o trabalho
 * para avaliar o que aconteceu
 * e como se pode melhorar

Análise Contínua
1. Kanban
 * Sistema de abastecimento e controle de fluxos produção
2. Heijunka
 * Ato de nivelar a variedade ou o volume de itens produzidos em um processo ao longo de um período de tempo
3. PDCA
 * Planejar
 * Executar
 * Verificar
 * Ajustar
4. Kaizen
 * atividades que melhorem continuamente todas as funções e envolvam todos os funcionários
5. Poka-Yoke
 * processo a prova de erros e falhas, destinado evitar defeitos na fabricação de produtos e serviços

Características
1. Processos “pull” (puxar)
 * em vez dos produtos serem “empurrados” para o fim da cadeia de produção
 * são “puxados” pelo cliente final, ouvindo e priorizando o que traz valor de negócio
2. Qualidade
 * sempre buscar detecção e solução dos problemas na sua origem, com intuito de não ter defeitos
3. Desperdício
 * eliminar atividades que
  * não têm valor
  * aumenta o trabalho
  * traz retrabalho
  * cria contra-tempos (pessoas, espaços, ferramentas, ambientes, documentos)
4. Melhoria Contínua
 * inspecionar
 * medir
 * encontrar as falhas
 * corrigir o quanto antes para aumentar produtividade
5. Manutenção
 * cultivar um relacionamento saudável com os fornecedores
 * para ter melhores negociações
 * com custos, informações e possíveis riscos que podem ocorrer
6. Flexibilidade
 * através de diferentes células
 * produz com velocidade variedades de produtos
 * sem comprometer a qualidade

1.   Scrum e XP são metodologias ágeis usadas no contexto de desenvolvimento de software que:
* Podem ser usadas em conjunto, criando uma metodologia híbrida própria, de forma a unir as práticas de gestão do Scrum com as práticas de desenvolvimento da XP.
2.   Comparando-se os métodos ágeis com os tradicionais, considere:

I.	Os métodos ágeis priorizam mais a implementação, e os tradicionais, geralmente, preocupam-se mais com a documentação.

II.	As metodologias ágeis são mais preditivas, e as tradicionais, mais adaptativas, ou seja, buscam se adaptar às mudanças ao invés de realizar longas análises sobre o futuro.

III.	Os métodos ágeis podem ser indicados quando o cliente não sabe muito bem o que deseja. Realizam entregas constantes que nem sempre acontecem nos métodos tradicionais, e os testes podem melhorar a qualidade do produto.

IV.	Os métodos ágeis não rejeitam processos e ferramentas dos tradicionais, mas as colocam em posição secundária, privilegiando os indivíduos e a comunicação entre eles.

São corretas:

* I, III e IV

7. Os modelos ágeis de desenvolvimento de software têm menos ênfase nas definições de atividades e mais ênfase na pragmática e nos fatores humanos do desenvolvimento. Um destes modelos enfatiza o uso de orientação a objetos e possui apenas duas grandes fases: 1 − Concepção e Planejamento e 2 − Construção. A fase de Concepção e Planejamento possui três disciplinas (chamadas de processos): Desenvolver Modelo Abrangente, Construir Lista de Funcionalidades e Planejar por funcionalidade. A fase de Construção incorpora duas disciplinas (processos): Detalhar por Funcionalidade e Construir por Funcionalidade.

O texto acima apresenta a metodologia ágil conhecida como:
* FDD (não Scrum)

10. Há diversos métodos que podem ser utilizados na construção de sistemas de informação. Sobre eles, analise:

I.	O desenvolvimento é feito em estágios formais, que devem evoluir em sequência e ter resultados definidos. Cada um precisa ser formalmente aprovado antes que o próximo se inicie. É indicado para grandes projetos que exijam especificações formais e rígido controle administrativo sobre cada estágio do desenvolvimento.

II.	Consiste em desenvolver um sistema experimental de maneira rápida e barata para que os usuários finais interajam com ele e o avaliem. Esse sistema é refinado e aperfeiçoado até que os usuários sintam que ele atende às suas necessidades, podendo ser usado como modelo para criar o sistema final.

Os itens I e II se referem, respectivamente:

* Ao ciclo de vida de sistema tradicional e à prototipagem. (não Aos modelos prescritivos e ágeis)
