# Linear Regression

````
# Linear Regression 1 Variable
# Independent: Number of A grades in 1st year (freshmen year)
x = [3,1,0,4]
# Dependent: Number of A grades in 2nd year (sophomore year)
y = [2,2,1,3]
# Hypothesis function
def hypothesis(theta0,theta1,x):
  h = theta0 + theta1 * x
  return h

# Cost function
def cost(theta0,theta1,x,y):
  # Number of training examples
  m = len(x)
  J = 0
  for i in range(m):
    J = J + ( hypothesis(theta0,theta1,x[i]) - y[i] )**2
    # print(i, " : ", J)
  J = J / (2*m)
  return J

theta0 = 0
theta1 = 1
J = cost(theta0,theta1,x,y)
print(J)

theta0 = -1
theta1 = 2
theta1 = 0.5
x = 6
x = 4
h = hypothesis(theta0,theta1,x)
print(h)

x = [1,2,4,0]
y = [0.5,1,2,0]
v_theta0 = [1,1,0.5,0,0.5]
v_theta1 = [0.5,1,0,0.5,0.5]
for i in range(len(v_theta0)):
  J = cost(v_theta0[i],v_theta1[i],x,y)
  print(i, " : ", J)
````

## Logistic Regression
One desires that
$$ 0 \leq h_\theta(x) \leq 1 $$

We have the sigmoid or logistic function
$$ g(z) = \frac{1}{1+e^{-z}} $$ 

With the following properties
\begin{align*}
g(z) \geq 0.5, \qquad  \text{when} \; z \geq 0
\end{align*}
\begin{align*}
&z=0, \qquad\quad e^{0}=1 &\Rightarrow g(z)=\frac{1}{2} \newline
&z \to \infty,\qquad\; e^{-\infty} \to 0 &\Rightarrow g(z)=1 \newline
&z \to -\infty,\quad\;\; e^{\infty}\to \infty &\Rightarrow g(z)=0
\end{align*}


Hypothesis representation
$$ h_\theta(x) = g(\theta^Tx) = \frac{1}{1+e^{-\theta^Tx}} $$ 

Thus
\begin{align*}
h_\theta(x) = g(\theta^T x) \geq 0.5, \qquad \text{when} \; \theta^T x \geq 0
\end{align*}
In a nutshell
\begin{align*}& \theta^T x \geq 0 \Rightarrow y = 1 \newline& \theta^T x < 0 \Rightarrow y = 0 \newline\end{align*}

Probability that $y=1$, given $x$ parametrized by $\theta$

$$h_\theta(x) = P(y=1|x;\theta) $$

Regarding that
$$  P(y=0) + P(y=1) = 1 $$

### Decision Boundary

if $ h_\theta(x) \geq 0.5 $,
predict $ y = 1$;
$ g(z) \geq 0.5 $,
when $z = \theta^Tx \geq 0$

if $ h_\theta(x) \lt 0.5 $,
predict $ y = 0$
$ g(z) \lt 0.5 $,
when $z = \theta^Tx \lt 0$

Decision boundary is a property of the hypothesis and not of the dataset

### Cost function

\begin{align*}
& J(\theta) = \dfrac{1}{m} \sum_{i=1}^m \mathrm{Cost}(h_\theta(x^{(i)}),y^{(i)}) \newline
& \mathrm{Cost}(h_\theta(x),y) = -\log(h_\theta(x)) \; & \text{if y = 1} \newline
& \mathrm{Cost}(h_\theta(x),y) = -\log(1-h_\theta(x)) \; & \text{if y = 0}
\end{align*}

\begin{align*}
& \mathrm{Cost}(h_\theta(x),y) = 0 \text{ if } h_\theta(x) = y \newline
& \mathrm{Cost}(h_\theta(x),y) \rightarrow \infty \text{ if } y = 0 \; \mathrm{and} \; h_\theta(x) \rightarrow 1 \newline
& \mathrm{Cost}(h_\theta(x),y) \rightarrow \infty \text{ if } y = 1 \; \mathrm{and} \; h_\theta(x) \rightarrow 0
\end{align*}

Penalizes absence of extremely high and low probability cases (as 0 and 1)

# Overfitting

Underfit: high bias

Overfit:
*   high variance
*   too many features
*   fit the training set very well
*   fail to new generalized cases

Prevent overfitting
1.   Reduce number of features
*   Manual selection
*   Algorithm selection

2.   Regularization
*   Keep all features, but reduce magnitude of parameters $\theta_j$
*   Work well, when a lot of slightly useful features
* Too large $\lambda$ can cause underfit
* $J(\theta)$ is still convex
* can impair the performance on training examples
* $\geq$ performance on predictions not in the training set


## Regularization





Cost function

Penalization

$$ \min_\theta = 
\dfrac{1}{2m} \sum_{i=1}^m (h_\theta(x^{(i)})-y^{(i)})^2 
+ \lambda \sum_{j=1}^n \theta_j^2 $$


### Gradient descent

\begin{align*} & \text{Repeat}\ \lbrace \newline & \ \ \ \ \theta_0 := \theta_0 - \alpha\ \frac{1}{m}\ \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})x_0^{(i)} \newline & \ \ \ \ \theta_j := \theta_j - \alpha\ \left[ \left( \frac{1}{m}\ \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)} \right) + \frac{\lambda}{m}\theta_j \right] &\ \ \ \ \ \ \ \ \ \ j \in \lbrace 1,2...n\rbrace\newline & \rbrace \end{align*}




$$ \theta_j := \theta_j \left(1- \alpha \frac{\lambda}{n}\right) 
- \frac{\alpha}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} $$


$\lambda, \alpha, m > 0 \implies 1- \alpha \frac{\lambda}{m} < 1 $ 

### Normal equation

\begin{align*}& \theta = \left( X^TX + \lambda \cdot L \right)^{-1} X^Ty \newline& \text{where}\ \ L = \begin{bmatrix} 0 & & & & \newline & 1 & & & \newline & & 1 & & \newline & & & \ddots & \newline & & & & 1 \newline\end{bmatrix}\end{align*}

$X$ is non-invertible if $m < n$, and may be non-invertible if $m = n$.

However $X^TX + \lambda \cdot L$ becomes invertible.

# Linear vs Logistic Regression

## Differences
\begin{align*}
&\text{Regression} & \text{Linear} & & \text{Logistic}\newline
&h = & \theta^T X && \dfrac{1}{1+e^{-\theta^T X}} \newline
&J = & \frac{(h-y)^T(h-y)}{2m} &&\frac{\left(-y^{T}\log(h)-(1-y)^{T}\log(1-h)\right)}{m}
\end{align*}

## Similarities
Gradient
$$\frac{\partial J}{\partial \theta}=  \frac{X^{T}(h-y)}{m} $$
Cost Function Penalty
$$pJ = \frac{\lambda}{2m} \sum_{j=1}^n \theta_j^2$$
Gradient Function Penalty
$$p\frac{\partial J}{\partial \theta} = \frac{\lambda}{m} \sum_{j=1}^n \theta_j$$

Remark
$\theta_0$ is not penalized

## Backpropagation
* $L$ = total number of layers in the network
* $s_l$ = number of units (not counting bias unit) in layer $l$
* $K$ = number of output units/classes

\begin{gather*}
J(\Theta) =
 - \frac{1}{m} \sum_{i=1}^m \sum_{k=1}^K 
 \left[y^{(i)}_k \log ((h_\Theta (x^{(i)}))_k)
 + (1 - y^{(i)}_k)\log (1 - (h_\Theta(x^{(i)}))_k)\right] 
 + \frac{\lambda}{2m}\sum_{l=1}^{L-1} \sum_{i=1}^{s_l} \sum_{j=1}^{s_{l+1}}
 ( \Theta_{j,i}^{(l)})^2
 \end{gather*}



Backpropagation

Given training set
$\{(x^{(1)},y^{(1)}),\cdot\cdot\cdot,(x^{(m)},y^{(m)})\}$
* Set 
$$\Delta_{ij}^{(l)}:=0 \quad \forall (l,i,j)$$

For training example $t=1:m$:


1.   Set 
$$a^{(1)}:= x^{(t)} \quad (\text{add } a_0^{(1)})$$
2.   Perform forward propagation to compute $a^{(l)}$ for $l=2,\cdot\cdot\cdot,L$
\begin{align}
z^{(2)} &= \Theta^{(1)}a^{(1)}\\
a^{(2)} &= g(z^{(2)}) \quad (\text{add } a_0^{(2)}) \\
z^{(3)} &= \Theta^{(2)}a^{(2)}\\
a^{(3)} &= g(z^{(3)}) \quad (\text{add } a_0^{(3)}) \\
z^{(4)} &= \Theta^{(3)}a^{(3)}\\
a^{(4)} &= g(z^{(4)})
\end{align}
3. Compute  the error
$$\delta^{(L)} = a^{(L)} - y^{(t)}$$
4. Compute $\delta^{(L-1)},\delta^{(L-2)},\cdot\cdot\cdot,\delta^{(2)}$
$$ \delta^{(l)} = ((\Theta^{(l)})^T\delta^{(l-1)}) \;.*\;
a^{(l)} \;.*\; (1-a^{(l)}) $$
where
$$ g'(z^{(l)}) = a^{(l)} \;.*\; (1-a^{(l)}) $$
5. Update $$ \Delta^{(l)}:= \Delta^{(l)} + \delta^{(l+1)}(a^{(l)})^T $$
6. Compute the derivatives
\begin{align}
D_{ij}^{(l)} &:= \dfrac{1}{m}\left(\Delta_{ij}^{(l)} + \lambda\Theta_{ij}^{(l)}\right) &\text{if } j \neq 0 \\
D_{ij}^{(l)} &:= \dfrac{1}{m}\left(\Delta_{ij}^{(l)}\right) 
&\text{if } j = 0
\end{align}
where
$$ \dfrac{\partial J(\Theta)}{\partial \Theta_{ij}^{(l)}} = D_{ij}^{(l)}
$$

## Kernels

Non-linear decision boundary

Gaussian

Given $x$, compute new feature $f_i$ depending on proximity to landmarks $l$ (new features)

\begin{equation*}
f_1 =\text{similarity}(x,l^{(1)})
= \exp\left(-\frac{\left\Vert x-l^{(1)} \right\Vert^2}{2\sigma^2}\right)
\end{equation*}

$sim(x,l^{(1)}) \in [0,1]$

If $ x = l^{(1)} $:
\begin{equation*}
f_1 
= \exp\left(-\frac{0^2}{2\sigma^2}\right)
= 1
\end{equation*}


If $ x \approx l^{(1)} $:
\begin{equation*}
f_1 
\approx \exp\left(-\frac{0^2}{2\sigma^2}\right)
\approx 1
\end{equation*}

If $x$ far from $l^{(1)}$:
\begin{equation*}
f_1 
\approx \exp\left(-\frac{\left(\text{large number}\right)^2}{2\sigma^2}\right)
\approx 0
\end{equation*}


* C: regularization parameter (large: all data must fit, including outlier, overfit)
* $\sigma$ bandwith how fast the similarity metric decreases (to 0)

* Large C
 * lower bias
 * higher variance (overfit)
 * ~ small $\lambda$
* Small C
 * higher bias (underfit)
 * lower variance
 * ~ large $\lambda$

* Large $\sigma^2$
 * features $f_i$ vary more smoothly
 * higher bias
 * lower variance

* Small $\sigma^2$
 * features $f_i$ vary less smoothly
 * lower bias
 * higher variance


* Overfit
 * Decrease C
 * Increase $\sigma^2$

* Underfit
 * Increase C
 * Decrease $\sigma^2$


Options for kernels
1. No kernels (Linear Kernel)
 * n (feature) large, m (train set) small

2. Gaussian Kernel
 * need to choose $\sigma^2$
 * n small and/or m large
 * perform feature scaling

# Logistic regression vs SVMs
* $n$: number of features
* $m$: number of training examples

If $n \gg m $ ($n=10.000$, $m=10\cdot\cdot\cdot1.000$)
* Logistic regression
* SVM without a kernel

If $n$ is small, $m$ is intermediate ($n=1\cdot\cdot\cdot1.000$, $m=10\cdot\cdot\cdot10.000$)
 * SVM with Gaussian kernel

If $n \ll m $ ($n=1\cdot\cdot\cdot1.000$, $m=50.000+$)
* Create/add more features
 * Logistic regression
 * SVM without a kernel

Neural network likely to work well for most of these settings, but may be slower to train

Logistic regression and SVM without a kernel
* provides a linear straight line decision boundary

$n=10$ features and $m=5000$ examples
* logistic regression classifier with gradient descent
 * training set: underfit (reduced performance on training or cross-validation sets)

* create/ add new polynomial features
* logistic regression cost function is convex, so gradient descent will always find the global minimum
